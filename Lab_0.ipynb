### импорт необходимых модулей

```python
%matplotlib inline
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

### считывание с вебки 

```python
vid = cv2.VideoCapture(0)
while True:
    """
        риал-тайм запись с вебки в отдельном окне. "Пробел" - сохраниение скриншота, "Esc" - выход
    """
    _, im = vid.read()
    blured = cv2.GaussianBlur(im,(0,0),5) #на правах анонимности
    cv2.imshow('test test',blured)
    k = cv2.waitKey(10)
    if k==27:
        vid.release()
        break
    if k==32:
        cv2.imwrite('test_image.jpg',blured)
cv2.destroyAllWindows()        
    
```

### вывод размытого жипега 

```python
scrshot = cv2.imread('test_image.jpg')
#так как opencv распаковывает в формате BGR, то надо поставить все на свои места
plt.imshow(cv2.cvtColor(scrshot,cv2.COLOR_BGR2RGB)), plt.axis("off")
plt.show()
```

```python
x=[0,100,200,200,100]
y=[100,0,0,300,300]
print(list((x,y) for x,y in zip(x,y)))
```

### рисуем звёздочки, линии и квадратики 

```python
plt.imshow(cv2.cvtColor(scrshot,cv2.COLOR_BGR2RGB)),plt.plot(x,y,'r*')
plt.plot(x[:2],y[:2])
plt.fill(x[1:],y[1:],facecolor='blue')
plt.show()
```
